```python
import numpy as np
from metric_learn import LMNN
from sklearn.datasets import load_iris

# Load Iris dataset
iris_data = load_iris()
X = iris_data['data']
Y = iris_data['target']

# Initialize LMNN with appropriate parameters
lmnn = LMNN(n_neighbors=5, learn_rate=1e-3)

# Fit the model
lmnn.fit(X, Y)
X_transformed = lmnn.transform(X)
print("Transformed data shape:", X_transformed.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# k-NN on original data
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)
acc_original = accuracy_score(Y_test, knn.predict(X_test))

# Transform data
X_train_transformed = lmnn.transform(X_train)
X_test_transformed = lmnn.transform(X_test)

# k-NN on transformed data
knn.fit(X_train_transformed, Y_train)
acc_transformed = accuracy_score(Y_test, knn.predict(X_test_transformed))

print("Accuracy on original data:", acc_original)
print("Accuracy on transformed data:", acc_transformed)
```

```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_classification(n_samples=200, n_features=2, n_classes=2, n_informative=2, n_redundant=0, random_state=42)
X[:, 1] *= 1000  # Scale the second feature to make it much larger

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title("Dataset with Non-Ideal Feature Scales")
plt.xlabel("Feature 1 (0-1)")
plt.ylabel("Feature 2 (0-1000)")
plt.show()
```

```python
from metric_learn import LMNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# k-NN without LMNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
acc_original = accuracy_score(y_test, knn.predict(X_test))

# Apply LMNN and re-train k-NN
lmnn = LMNN(learn_rate=0.001, n_neighbors=5)
lmnn.fit(X_train, y_train)
X_train_lmnn = lmnn.transform(X_train)
X_test_lmnn = lmnn.transform(X_test)

knn.fit(X_train_lmnn, y_train)
acc_transformed = accuracy_score(y_test, knn.predict(X_test_lmnn))

print("Accuracy without LMNN:", acc_original)
print("Accuracy with LMNN:", acc_transformed)
```
